```python
#import libraries that we'll use
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#extract our data in a csv file and shows the head of data
df = pd.read_csv("https://raw.githubusercontent.com/sushantag9/Supermarket-Sales-Data-Analysis/master/supermarket_sales%20-%20Sheet1.csv")
df.head()
```

```python
#look through the columns we have and the data entries
df.info()
```

```python
#merge date and time to make it easier to analyze
df["DateTime"] = pd.to_datetime(df["Date"] + " " + df["Time"])
df
```

```python
#see our merge result in columns info
df.info()
```

```python
#drop the individual date and time columns since we have the merged one
df.drop(columns=["Date","Time"],inplace=True)
df.info()
```

### Line Plot

```python
#extract hour, weekday, and month from date time for further analyze
df["hour"] = df["DateTime"].dt.hour
df["weekday"] = df["DateTime"].dt.weekday
df["month"] = df["DateTime"].dt.month
```

```python
#shows the extract result in the data head
df.head()
```

```python
# get our data quantity in each section
df_plot_hour = df.groupby("hour").agg({"Quantity":"sum"}).reset_index()
df_plot_weekday = df.groupby("weekday").agg({"Quantity":"sum"}).reset_index()
df_plot_month = df.groupby("month").agg({"Quantity":"sum"}).reset_index()
```

shows respective column and their quantity

```python
df_plot_hour
```

```python
df_plot_weekday
```

```python
df_plot_month
```

project the visual plot in line plot

```python
sns.lineplot(data=df_plot_hour,x="hour",y="Quantity")
plt.title("Quantity in Hour")
plt.show()
```

```python
sns.lineplot(data=df_plot_weekday,x="weekday",y="Quantity")
plt.title("Quantity in Weekday")
plt.show()
```

```python
sns.lineplot(data=df_plot_month,x="month",y="Quantity")
plt.title("Quantity in Month")
plt.show()
```
