### Data Preparation

```python
# Import libraries that we'll use
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import our dataset from a csv file and shows the first 5 row of data
df = pd.read_csv("https://raw.githubusercontent.com/sushantag9/Supermarket-Sales-Data-Analysis/master/supermarket_sales%20-%20Sheet1.csv")
df.head()
```

```python
# look through the columns we have and the data entries
df.info()
```

```python
# Merge date and time to make it easier to analyze
df["DateTime"] = pd.to_datetime(df["Date"] + " " + df["Time"])
```

```python
# See our merge result in columns info
df.info()
```

```python
# Clean the data by dropped redundant date and time columns
df.drop(columns=["Date","Time"],inplace=True)
df.info()
```

```python
# Extract weekday from date time for further analyze
df["weekday"] = df["DateTime"].dt.weekday
```

```python
# Get our weekday data quantity
df_plot_weekday = df.groupby("weekday").agg({"Quantity":"sum"}).reset_index()
df_plot_weekday
```

## Line Plot



```python
sns.lineplot(data=df_plot_weekday,x="weekday",y="Quantity")
plt.title("Quantity in Weekday")
plt.show()
```

## Histograms

```python
sns.histplot(df,x="Rating")
```

## Box Plot

```python
sns.boxplot(data=df,y="Total")
```
